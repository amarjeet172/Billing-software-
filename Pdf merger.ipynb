```python
import os
import tkinter as tk
from tkinter import filedialog, messagebox
from PyPDF2 import PdfReader, PdfWriter
from PIL import Image

def is_blank(page):
    """Check if a page is blank using Pillow."""
    img = Image.open(page)
    img = img.convert('L')  # Convert to grayscale
    img_data = img.getdata()
    if img_data.count(255) > 0.98 * len(img_data):  # 98% white pixels considered blank
        return True
    return False

def remove_blank_pages(input_pdf_path, output_pdf_path):
    """Remove blank pages from a PDF."""
    pdf_reader = PdfReader(input_pdf_path)
    pdf_writer = PdfWriter()

    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        if not is_blank(page):
            pdf_writer.add_page(page)

    with open(output_pdf_path, 'wb') as out_pdf_file:
        pdf_writer.write(out_pdf_file)

def merge_pdfs(folder_path, merged_pdf_path):
    """Merge all PDFs in a folder into one."""
    pdf_writer = PdfWriter()

    for filename in os.listdir(folder_path):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(folder_path, filename)
            pdf_reader = PdfReader(pdf_path)
            for page in pdf_reader.pages:
                pdf_writer.add_page(page)

    with open(merged_pdf_path, 'wb') as out_pdf_file:
        pdf_writer.write(out_pdf_file)

def select_folder():
    """Select folder using GUI."""
    folder_path = filedialog.askdirectory()
    if folder_path:
        merged_pdf_path = os.path.join(folder_path, 'merged.pdf')
        merge_pdfs(folder_path, merged_pdf_path)
        remove_blank_pages(merged_pdf_path, merged_pdf_path)
        messagebox.showinfo('Success', 'PDFs merged and blank pages removed successfully!')

# Create GUI
root = tk.Tk()
root.title('PDF Merger & Blank Page Remover')

frame = tk.Frame(root)
frame.pack(padx=20, pady=20)

select_button = tk.Button(frame, text='Select Folder', command=select_folder)
select_button.pack()

root.mainloop()
```

```python
import os
import tkinter as tk
from tkinter import filedialog, messagebox
from PyPDF2 import PdfReader, PdfWriter
import fitz  # PyMuPDF
from PIL import Image

def is_blank(image_path):
    """Check if an image is blank using Pillow."""
    img = Image.open(image_path)
    img = img.convert('L')  # Convert to grayscale
    img_data = img.getdata()
    if img_data.count(255) > 0.98 * len(img_data):  # 98% white pixels considered blank
        return True
    return False

def convert_pdf_to_images(input_pdf_path, output_folder):
    """Convert each page of a PDF into an image."""
    images = []
    pdf_document = fitz.open(input_pdf_path)
    
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        pix = page.get_pixmap()
        image_path = os.path.join(output_folder, f'page_{page_num}.png')
        pix.save(image_path)
        images.append(image_path)
    
    return images

def remove_blank_pages(input_pdf_path, output_pdf_path):
    """Remove blank pages from a PDF."""
    image_paths = convert_pdf_to_images(input_pdf_path, os.path.dirname(input_pdf_path))
    
    pdf_writer = PdfWriter()
    for image_path in image_paths:
        if not is_blank(image_path):
            with open(image_path, 'rb') as img_file:
                img_data = img_file.read()
            pdf_writer.add_page(img_data)
            os.remove(image_path)
    
    with open(output_pdf_path, 'wb') as out_pdf_file:
        pdf_writer.write(out_pdf_file)

def merge_pdfs(folder_path, merged_pdf_path):
    """Merge all PDFs in a folder into one."""
    pdf_writer = PdfWriter()

    for filename in os.listdir(folder_path):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(folder_path, filename)
            pdf_reader = PdfReader(pdf_path)
            for page in pdf_reader.pages:
                pdf_writer.add_page(page)

    with open(merged_pdf_path, 'wb') as out_pdf_file:
        pdf_writer.write(out_pdf_file)

def select_folder():
    """Select folder using GUI."""
    folder_path = filedialog.askdirectory()
    if folder_path:
        merged_pdf_path = os.path.join(folder_path, 'merged.pdf')
        merge_pdfs(folder_path, merged_pdf_path)
        remove_blank_pages(merged_pdf_path, merged_pdf_path)
        messagebox.showinfo('Success', 'PDFs merged and blank pages removed successfully!')

# Create GUI
root = tk.Tk()
root.title('PDF Merger & Blank Page Remover')

frame = tk.Frame(root)
frame.pack(padx=20, pady=20)

select_button = tk.Button(frame, text='Select Folder', command=select_folder)
select_button.pack()

root.mainloop()
```
