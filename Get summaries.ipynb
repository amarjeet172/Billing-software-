# GET BILLS RECORDS

```python
import os
import shutil
import re
from openpyxl import load_workbook
from datetime import datetime
from win32com import client
from datetime import date

ParentFolder = r"C:\Users\AMARJEET\Desktop\python_bill\AUDIT"
ListOfAllFiles= os.listdir(ParentFolder)
#print(ListOfAllFiles)
today=str(date.today())


source = r"C:\Users\AMARJEET\Desktop\python_bill\StockRemFormate.xlsx"
destination= r"C:\Users\AMARJEET\Desktop\python_bill\Summary\Summary_1"+today+".xlsx"
file_loc=r"C:\Users\AMARJEET\python_bill\CUSTOMER.txt"
dest = shutil.copyfile(source, destination)
    #load excel file
workbook = load_workbook(filename=destination)
    #open workbook
sheet = workbook.active
write_index = 4
for excels in ListOfAllFiles:
    if ".xlsx" in excels and excels[0]=='A':
        pur_destination=ParentFolder+'/'+excels
    
        #load excel file
        workbook_pur = load_workbook(filename=pur_destination) 
        sheet_pur = workbook_pur.active
        print(sheet_pur)
        for i in range(6,100):
            cell_str="V"+str(i)
            try:
                Stock=int(sheet_pur[cell_str].value)
            except TypeError:
                break  
            if Stock !=0:
                BillNo = sheet_pur["M3"].value
                BillDate = sheet_pur["C3"].value
                cell_str="R"+str(i)
                slNo=int(sheet_pur[cell_str].value)
                quantity =Stock
                cell_str="S"+str(i)
                GST=int(sheet_pur[cell_str].value)
                cell_str="I"+str(i)
                rate=round(float(sheet_pur[cell_str].value),2)
                cell_str="J"+str(i)
                dis=round(float(sheet_pur[cell_str].value),2)
                rate=round((rate+(0.032*rate)),2)       
                amount=(quantity*rate)-(quantity*rate*dis/100)
                NetAmmount=amount+(amount*GST/100)
                
                cur_cell_str = "A"+str(write_index)
                sheet[cur_cell_str]=BillNo
               
                cur_cell_str = "B"+str(write_index)
                sheet[cur_cell_str]=BillDate
                
                cur_cell_str = "C"+str(write_index)
                sheet[cur_cell_str]=slNo
                
                cur_cell_str = "D"+str(write_index)
                sheet[cur_cell_str]=Stock
                
                
                cur_cell_str = "E"+str(write_index)
                sheet[cur_cell_str]=GST
                
                cur_cell_str = "F"+str(write_index)
                sheet[cur_cell_str]=NetAmmount
                write_index = write_index+1
        workbook_pur.save(filename=pur_destination)
                    
                    
                    
workbook.save(filename=destination)
                
    
                
        

            
```
